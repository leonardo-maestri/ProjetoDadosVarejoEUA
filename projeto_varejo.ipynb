## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
df5 = df.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by= 'Valor_Venda', ascending = False)

# Função para converter os dados em valor absoluto pq estão em notação cientifica 
def autopct_format(values): 
    def my_format(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format

df5.head()
```

```python
# plot gráfico em pizza

# Tamanho da figura
plt.figure(figsize = (16, 6))

# Gráfico de pizza
plt.pie(df5['Valor_Venda'], 
        labels = df5['Segmento'],
        autopct = autopct_format(df5['Valor_Venda']), # usando a função pra converter os valores das vendas
        startangle = 90)

# Limpa o círculo central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotações
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df5['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

## Pergunta de Negócio 7 :

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
# função para aplicar o desconto
def desconto(valor_venda):
    if valor_venda > 1000:
        valor_venda = valor_venda * 0.85
        return valor_venda
    elif valor_venda < 1000:
        valor_venda = valor_venda * 0.90
        return valor_venda
# criando nova coluna, aplicando os valores com desconto com map
df['Vendas_Desconto'] = df['Valor_Venda'].map(desconto)

# variável pra saber a quantidade de vendas que tiveram descontos de 15%
vendas_com_desconto15 = df[df['Valor_Venda'] > 1000].shape[0]


print('A quantidade de vendas que tiveram desconto é:', vendas_com_desconto15)
df.head(5)
    
```
